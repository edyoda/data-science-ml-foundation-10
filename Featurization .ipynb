```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
titanic_data = pd.read_csv('titanic.csv')
```

```python
titanic_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
#creates a mapping or logic that is capable of transformation
le.fit(titanic_data['Sex'])
```

```python
le.transform(titanic_data['Sex'])
```

```python
titanic_data['Sex'] = le.fit_transform(titanic_data['Sex'])
```

```python
titanic_data.head()
```

```python
del titanic_data['Sex-t']
```
