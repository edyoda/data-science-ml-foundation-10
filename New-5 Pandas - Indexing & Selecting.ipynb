### iloc - Integer Oriented Indexing

```python
dubaiPrices
```

```python
dubaiPrices.iloc[1:,2:]
```

```python
#Accessing by index id
dubaiPrices.iat[3,0]
```

```python
#Accessing by value
dubaiPrices.at['2013-Q4','Dubai']
```

### Thumb Rules
* at -> Access a specific value given index & col
  dubaiPrices.at['2013-Q4','Dubai']
  
* iat -> Access a specific value given index id & col id
* loc -> Access row by index value 
  NYC_SnowAvgs_index.loc['January']['Avg SnowDays']
  
* iloc -> Access row by index id
