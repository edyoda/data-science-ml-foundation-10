```python
COLUMNS = ["crim", "zn", "indus", "nox", "rm", "age",
           "dis", "tax", "ptratio", "medv"]
```

```python
FEATURES = ["crim", "zn", "indus", "nox", "rm",
            "age", "dis", "tax", "ptratio"]
LABEL = "medv"
```

```python
training_set = pd.read_csv(r"D:\Data-Science-ML\data-science\boston_train.csv", skipinitialspace=True,
                           skiprows=1, names=COLUMNS)
test_set = pd.read_csv(r"D:\Data-Science-ML\data-science\boston_test.csv", skipinitialspace=True,
                       skiprows=1, names=COLUMNS)
prediction_set = pd.read_csv(r"D:\Data-Science-ML\data-science\boston_predict.csv", skipinitialspace=True,
                             skiprows=1, names=COLUMNS)
```

```python
feature_cols = [tf.contrib.layers.real_valued_column(k)
                  for k in FEATURES]
```

```python
regressor = tf.contrib.learn.DNNRegressor(feature_columns=feature_cols,
                                          hidden_units=[10, 10],
                                          model_dir=r"D:\Data-Science-ML\data-science\boston_model")
```

```python
def input_fn(data_set):
    feature_cols = {k: tf.constant(data_set[k].values)
                  for k in FEATURES}
    labels = tf.constant(data_set[LABEL].values)
    return feature_cols, labels
```

```python
regressor.fit(input_fn=lambda: input_fn(training_set), steps=5000)
```

```python
ev = regressor.evaluate(input_fn=lambda: input_fn(test_set), steps=1)
```

```python
loss_score = ev["loss"]
print("Loss: {0:f}".format(loss_score))
```

```python
import itertools
y = regressor.predict(input_fn=lambda: input_fn(prediction_set))
# .predict() returns an iterator; convert to a list and print predictions
predictions = list(itertools.islice(y, 6))
print ("Predictions: {}".format(str(predictions)))
```

### Convolutional Neural Networks
* Detecting puppy in pictures
* recognizing spoken words

* Humans do this effortlessly, because of visual, auditory & other sensory modules in our brains

##### Architecture of the Visual Cortex
* Neurons in visual cortex have a small local receptive field.
* Visual cortex identifies the corners
* Some neuros have larger fields & some smalls.
* Compare each neuron with a perceptron
* All these neurons are layered. Or known as Convolutional Layers or pooling layers

#### Convolutional Layer
* Neurons of 1st layer are not connected to every single pixel in the input image, but only to receptive fields.
* This architecture allows the network to concentrate low level features in first hidden layers & then assemble the output of low level layers into higher layers

```python
import numpy as np
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg")
flower = load_sample_image("flower.jpg")

image = china[150:220, 130:250]
height, width, channels = image.shape
image_grayscale = image.mean(axis=2).astype(np.float32)
images = image_grayscale.reshape(1, height, width, 1)  
```

```python
images.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot_image(image):
    plt.imshow(image, cmap="gray", interpolation="nearest")
    plt.axis("off")
```

```python
def plot_color_image(image):
    plt.imshow(image.astype(np.uint8),interpolation="nearest")
    plt.axis("off")
```

```python
plot_image(images[0,:,:,0])
```

```python
import tensorflow as tf
```

```python
fmap = np.zeros(shape=(7, 7, 1, 2), dtype=np.float32)
```

```python
fmap.shape
```

```python
fmap[:, 3, 0, 0] = 1
```

```python
fmap[3, :, 0, 1] = 1
```

```python
fmap[:, :, 0, 0]
```

```python
plot_image(fmap[:, :, 0, 0])
plt.show()
plot_image(fmap[:, :, 0, 1])
plt.show()
```

```python
tf.reset_default_graph()
```

```python
X = tf.placeholder(tf.float32, shape=(None, height, width, 1))
```

```python
feature_maps = tf.constant(fmap)
```

```python
convolution = tf.nn.conv2d(X, feature_maps, strides=[1,1,1,1], padding="SAME", use_cudnn_on_gpu=False)
```

```python
with tf.Session() as sess:
    output = convolution.eval(feed_dict={X: images})
```

```python
plot_image(images[0, :, :, 0])
#save_fig("china_original", tight_layout=False)
plt.show()
```

```python
plot_image(output[0, :, :, 0])
#save_fig("china_vertical", tight_layout=False)
plt.show()
```

```python
plot_image(output[0, :, :, 1])
#save_fig("china_horizontal", tight_layout=False)
plt.show()
```
