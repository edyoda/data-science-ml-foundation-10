```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_sample_images
```

```python
# Load sample images
dataset = np.array(load_sample_images().images, dtype=np.float32)
batch_size, height, width, channels = dataset.shape
```

```python
filters_test = np.zeros(shape=(7, 7, channels, 2), dtype=np.float32)
```

```python
filters_test[:, 3, :, 0] = 1  # vertical line
filters_test[3, :, :, 1] = 1  # horizontal line
```

```python
# Create a graph with input X plus a convolutional layer applying the 2 filters
X = tf.placeholder(tf.float32, shape=(None, height, width, channels))
```

```python
convolution = tf.nn.conv2d(X, filters_test, strides=[1,2,2,1], padding="SAME")
```

```python
with tf.Session() as sess:
    output = sess.run(convolution, feed_dict={X: dataset})
```

```python
plt.imshow(output[0, :, :, 1])  # plot 1st image's 2nd feature map
plt.show()
```

```python
import pandas as pd
```

```python
pd.read_csv(r'D:\Data-Science-ML\data-science\boston_train.csv')
```

```python
COLUMNS = ["crim", "zn", "indus", "nox", "rm", "age",
           "dis", "tax", "ptratio", "medv"]
```

```python
FEATURES = ["crim", "zn", "indus", "nox", "rm",
            "age", "dis", "tax", "ptratio"]
LABEL = "medv"
```

```python
training_set = pd.read_csv(r"D:\Data-Science-ML\data-science\boston_train.csv", skipinitialspace=True,
                           skiprows=1, names=COLUMNS)
test_set = pd.read_csv(r"D:\Data-Science-ML\data-science\boston_test.csv", skipinitialspace=True,
                       skiprows=1, names=COLUMNS)
prediction_set = pd.read_csv(r"D:\Data-Science-ML\data-science\boston_predict.csv", skipinitialspace=True,
                             skiprows=1, names=COLUMNS)
```

```python
feature_cols = [tf.contrib.layers.real_valued_column(k)
                  for k in FEATURES]
```

```python
regressor = tf.contrib.learn.DNNRegressor(feature_columns=feature_cols,
                                          hidden_units=[10, 10],
                                          model_dir=r"D:\Data-Science-ML\data-science\boston_model")
```

```python
def input_fn(data_set):
    feature_cols = {k: tf.constant(data_set[k].values)
                  for k in FEATURES}
    labels = tf.constant(data_set[LABEL].values)
    return feature_cols, labels
```

```python
regressor.fit(input_fn=lambda: input_fn(training_set), steps=5000)
```

```python
ev = regressor.evaluate(input_fn=lambda: input_fn(test_set), steps=1)
```

```python
loss_score = ev["loss"]
print("Loss: {0:f}".format(loss_score))
```
