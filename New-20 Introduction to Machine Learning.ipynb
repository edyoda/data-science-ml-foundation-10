#### Supervised Learning Example : Simple linear Regression

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
rng = np.random.RandomState(11)
x = 10 * rng.rand(50)
x
```

```python


x[:,np.newaxis].shape #add one more dim to data
#x.reshape((50,1))
#Another approach
x.reshape((50,1))
```

```python
y = 8*x + 145
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#create empty logic or brain
model = LinearRegression(fit_intercept=True)
```

```python
#Trained the model with feature & target
#Input data should be in form of list of rows
model.fit(x.reshape(50,1),y)
```

```python
x.reshape(50,1)
#or
x[:,np.newaxis]
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5.6181871]])
```
